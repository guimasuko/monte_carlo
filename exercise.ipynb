## Exercício

Marcelo recebeu 250000 reais de herança e deseja ficar um período sem trabalhar. Ele, por levar uma vida minimalista e já possuir um imóvel, tem gastos mensais que podem ser bem modelados por uma distribuição normal com média 3000 e desvio padrão 1000. Considere também que ele tem acesso a um investimento que rende líquido 1% ao mês.

* (a) Por quantos meses em média ele consegue viver sem trabalhar?

* (b) Plot a distribuição do número de meses que ele pode ficar sem trabalhar.

* (c)  Considere agora que o desvio padrão da sua distribuição de gastos aumenta para 2000. Como isso afeta os resultados?

Sugira uma métrica mais robusta que a média para informar o número de meses que ele pode ficar sem trabalhar.

Observação: Esse exercício pode ser implementado de duas formas, usando um loop while e usando uma recursão. Faça uma função para cada uma das formas em Python.

```python
# packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# hide warning messages
import warnings
warnings.filterwarnings("ignore")
```

```python
# better plots
sns.set(rc={'figure.figsize':(12,8)});
```

```python
def monte_carlo(n):
    lista_n = []
    for i in range(n):
        # Marcelo's wealth
        y = 250000

        # number of months without work
        number_months = 0

        # distribution size of Marcelo's spending variable
        while (y > 0):
            # Marcelo's spending variable
            x = np.random.normal(3000, 1000)

            # Marcelo's wealth update
            y = (y - x)*1.01

            # number of months without work update
            number_months = number_months + 1
        lista_n.append(number_months)

    # distribution plot
    plt.hist(lista_n, bins=50)
    plt.axvline(np.mean(lista_n), color='r', linestyle = '--', linewidth = 2)
    plt.xlabel('Meses sem trabalhar')
    plt.ylabel('Frequência')
    plt.title('Distribuição dos meses sem trabalhar com desvio-padrão de 1000')

    # number of months on average
    mean_number_months = np.mean(lista_n)

    return(mean_number_months)
```

```python
monte_carlo(100)
```

```python
monte_carlo(1000)
```

```python
monte_carlo(10000)
```

```python
def monte_carlo_(n):
    lista_n = []
    for i in range(n):
        # Marcelo's wealth
        y = 250000

        # number of months without work
        number_months = 0

        # distribution size of Marcelo's spending variable
        while (y > 0):
            # Marcelo's spending variable
            x = np.random.normal(3000, 2000)

            # Marcelo's wealth update
            y = (y - x)*1.01

            # number of months without work update
            number_months = number_months + 1
        lista_n.append(number_months)

    # distribution plot
    plt.hist(lista_n, bins=50)
    plt.axvline(np.mean(lista_n), color='r', linestyle = '--', linewidth = 2)
    plt.xlabel('Meses sem trabalhar')
    plt.ylabel('Frequência')
    plt.title('Distribuição dos meses sem trabalhar com desvio-padrão de 2000')

    # number of months on average
    mean_number_months = np.mean(lista_n)

    return(mean_number_months)
```
